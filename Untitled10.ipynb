<font color='red'>$ \\ \textbf{Edge detection:}  $</font>


<b>a) Sobel :</b><br> -Detects edges are where the gradient magnitude is high.This makes the Sobel edge detector more sensitive to diagonal edge than horizontal and vertical edges. <br>
-The operator uses two kernels which are convolved with the original image to calculate approximations of the derivatives (one for horizontal changes, and one for vertical). If we define A as the source image, and Gx and Gy are two images which at each point contain the vertical and horizontal derivative approximations.<br>-
At each point in the image, the resulting gradient approximations can be combined to give the gradient magnitude.<br>
-above process is in Image.py



```python
img1=my_image.readimage('images/PGN1.jpg')
img2=my_image.readimage('images/edgedetection/test1.png')
img3=my_image.readimage('images/test5.png')
img4=my_image.readimage('images/edgedetection/test2.png')
img5=my_image.readimage('images/edgedetection/test3.png')


img1_,edge1=my_image.edge_detection(img1,mod='sobel',blur_sigma=1)
img2_,edge2=my_image.edge_detection(img2,mod='sobel',blur_sigma=1)
img3_,edge3=my_image.edge_detection(img3,mod='sobel',blur_sigma=0.1)
img4_,edge4=my_image.edge_detection(img4,mod='sobel',blur_sigma=0.1)
img5_,edge5=my_image.edge_detection(img5,mod='sobel',blur_sigma=0.1)



Show.compareim(img1_,edge1,'','',size=2)
Show.compareim(img2_,edge2,'','',size=1.2)
Show.compareim(img3_,edge3,'','',size=2)
Show.compareim(img4_,edge4,'','',size=2)
Show.compareim(img5_,edge5,'','',size=2)
```
